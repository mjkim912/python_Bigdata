```python
import pandas as pd
from sklearn import svm, metrics
import random, re

# 데이터 불러오기
csv = []

with open ('iris.csv', 'r', encoding = 'utf-8') as fp :
    # 데이터를 한 줄 씩 읽어오기
    for line in fp :
        line = line.strip()      # 좌우 공백 제거
        cols = line.split(',')   # 필드(열)를 콤마기호로 구분
        
        # 문자열 데이터를 숫자로 변환
        fn = lambda n : float(n) if re.match(r'^[0-9\.]+$', n) else n
        cols = list(map(fn, cols))
        csv.append(cols)
        

# 헤더를 제거한다.
# del csv[0]

# 데이터 섞어주기(셔플)
random.shuffle(csv)


# -------------------------------------------------------------------------------------------------

# 학습전용 & 테스트전용으로 데이터를 2 : 1 비율로 조정 (전체 150개)
# 학습용 = 100개, 테스트용 = 50개

total_len = len(csv)  # 전체 개수
print('total : ', total_len)

train_len = int(total_len * 2/3)
print('train : ', train_len)

train_data = []
train_label = []
test_data = []
test_label = []

for i in range(total_len) :
    data = csv[i][0:4]
    label = csv[i][4]   # 정답이 있는 name 항목의 값을 저장
    
    # 데이터 구분(2/3)
    if i < train_len :
        train_data.append(data)    # 100개의 데이터 입력
        train_label.append(label)
    else :
        test_data.append(data)     # 50개의 데이터 입력
        test_label.append(label)
        
        
# --------------------------------------------------------------------------------------------------

# 컴퓨터 학습시킬 알고리즘 선택
clf = svm.SVC()

# 컴퓨터 학습
clf.fit(train_data, train_label)

# 결과예측
predict = clf.predict(test_data)
print("예측결과는 : ", predict)

# 정확도 출력
score = metrics.accuracy_score(predict, test_label)
print("정확도 : ", score)
```

- 텍스트 내용을 읽어온다.

```python
from sklearn import svm, metrics, model_selection 
                  # 분석기, 정확도, 보고서
import os.path, re, json
from glob import glob

files = glob('./lang/train/*.txt')

train_data = []
train_label = []

for file_name in files :
    basename = os.path.basename(file_name)
    lang = basename.split('-')[0]
    
    # 텍스트 추출하기
    file = open(file_name, 'r', encoding = 'utf-8')
    text = file.read()
    text = text.lower()  
    # 대문자와 소문자 섞여 있으면 분석을 두 번 한다. 속도를 빠르게 하기 위해 한가지 종류로 바꿔준다.
    file.close()
    
    
    # 알파벳 출연빈도
    code_a = ord('a')  # 97
    code_z = ord('z')  # 122
    
    # 알파벳 26자를 저장할 초기 변수 선언
    count = [0] * 26
    
    #index_a = (code_a - 가져온 글자ord('알파벳')) * -1
    #       97   - 98 (b)  = 1 (b의 count 인덱스)
    
    
    # 알파벳 출현빈도
    # 알파벳의 빈도수에 따라 언어가 달라진다.
    for character in text :
        code_current = ord(character)
        
        if code_a <= code_current <= code_z :    # 알파벳 소문자만 작업하겠다. 한글, 숫자 제외
            count[code_current - code_a] += 1
            
    #print(lang, count)
    
    # 정규화 작업 ( Normalization )
    # 데이터의 값이 크면 분포도가 넓어지기 때문에 표준화 시켜주어야 한다.
    # 데이터의 값을 0.0 ~ 1.0 사이의 값으로 변환시킨다.
    
    
    total = sum(count)                            # 전체 합계를 구한다. 
    count = list(map(lambda n: n/total, count))  # 각각의 값을 total로 나눈다.
    
    train_data.append(count)   # 학습용 데이터를 저장한다.
    train_label.append(lang)   # 학습용 정답을 저장한다.
    

    
# -----------------------------------------------------------------------------------------------------

files = glob('./lang/test/*.txt')

test_data = []
test_label = []

for file_name in files :
    basename = os.path.basename(file_name)
    lang = basename.split('-')[0]
    
    # 텍스트 추출하기
    file = open(file_name, 'r', encoding = 'utf-8')
    text = file.read()
    text = text.lower()  
    # 대문자와 소문자 섞여 있으면 분석을 두 번 한다. 속도를 빠르게 하기 위해 한가지 종류로 바꿔준다.
    file.close()
    
    
    # 알파벳 출연빈도
    code_a = ord('a')  # 97
    code_z = ord('z')  # 122
    
    # 알파벳 26자를 저장할 초기 변수 선언
    count = [0] * 26
    
    #index_a = (code_a - 가져온 글자ord('알파벳')) * -1
    #       97   - 98 (b)  = 1 (b의 count 인덱스)
    
    
    # 알파벳 출현빈도
    # 알파벳의 빈도수에 따라 언어가 달라진다.
    for character in text :
        code_current = ord(character)
        
        if code_a <= code_current <= code_z :    # 알파벳 소문자만 작업하겠다. 한글, 숫자 제외
            count[code_current - code_a] += 1
            
    #print(lang, count)
    
    # 정규화 작업 ( Normalization )
    # 데이터의 값이 크면 분포도가 넓어지기 때문에 표준화 시켜주어야 한다.
    # 데이터의 값을 0.0 ~ 1.0 사이의 값으로 변환시킨다.
    
    
    total = sum(count)                            # 전체 합계를 구한다. 
    count = list(map(lambda n: n/total, count))  # 각각의 값을 total로 나눈다.
    
    test_data.append(count)   # 테스트용 데이터를 저장한다.
    test_label.append(lang)   # 테스트용 정답을 저장한다.
    
    
# -----------------------------------------------------------------------------------------------------

clf = svm.SVC()

clf.fit(train_data, train_label)

predict = clf.predict(test_data)

score = metrics.accuracy_score(predict, test_label)
report = metrics.classification_report(predict, test_label)

print('정확도 : ', score)
print()
print()
print(report)
```
