```python
# 인구데이터가 101개가 맞는지 확인

import csv
f = open('pop.csv')
data = csv.reader(f)

# 데이터를 담을 리스트변수 초기화 선언
result = []

for row in data :
    if ('분당구 정자1동' in row[0]):            # 특정 지역을 검색
        print('데이터 길이 : ', len(row[3:]))
        
        for i in row[3:]:                       # 0 ~ 100세 이상의 모든 인구 찾기
            result.append(i)                     # 해당 연령의 인구 리스트에 저장
            
print()
print(result)
```

```python
# 작성된 데이터가 문자열이므로 정수형으로 변환한다.

import csv
f = open('pop.csv')
data = csv.reader(f)

# 데이터를 담을 리스트변수 초기화 선언
result = []

for row in data :
    if ('분당구 정자1동' in row[0]):            # 특정 지역을 검색
        print('데이터 길이 : ', len(row[3:]))
        
        for i in row[3:]:                       # 0 ~ 100세 이상의 모든 인구 찾기
            result.append(int(i))                # 해당 연령의 인구 리스트에 저장
            
print()
print(result)
```

# 우리동네 성별 분석

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

male = []
female = []

name = input('지역명 입력(읍면동 단위) : ')

for row in data :
    if name in row[0] :
        for i in range(0, 101) :
            #i = i.replace(',', '')           # 천단위가 없어서 에러가 남.
            
            male.append(int(row[i+3]))         # 3:103
            female.append(int(row[-(i+1)]))        # 끝에서부터
            

female.reverse()     # 데이터 거꾸로 정렬            
            
# ------------------------------------------------------------------------

print('남자 : ', len(male))
print('여자 : ', len(female))

plt.rc('font', family='Malgun Gothic')      # 맥 : Apple gothic  : 한글깨짐 방지
plt.title(name + '지역 성별 비율')

plt.barh(range(101), male, label='남자')
plt.barh(range(101), female, label='여자')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

male = []
female = []

name = input('지역명 입력(읍면동 단위) : ')

for row in data :
    if name in row[0] :
        for i in row[3:104] : 
            male.append(-int(i))
            
        for i in row[106:] :
            female.append(int(i))
            
            
# ------------------------------------------------------------------------

print('남자 : ', len(male))
print('여자 : ', len(female))

plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi = 300)        # 그래프 크기 선택(inch)

plt.rc('font', family='Malgun Gothic')        # 맥 : Apple gothic  : 한글깨짐 방지
plt.rcParams['axes.unicode_minus'] = False   # - 기호 깨짐 방지

plt.title(name + '지역 남여 성별 비율')

plt.barh(range(101), male, label='남자')
plt.barh(range(101), female, label='여자')

plt.legend()                                  # 범례 출력 (label 선언이 선행되어야 함)
plt.show()
```

## 제주도에는 실제로 여성의 비율이 많을까?

#### 제주특별자치도

- 남 : 338,975
- 여 : 337,104

제주특별자치도를 검색하면 전체 검색결과가 출력되는데, 그중에서 첫번째 데이터가 전체합계에 대한 정보가 들어 있다.

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

male = []
female = []
m = 0
f = 0

name = input('지역명 입력(읍면동 단위) : ')  # 제주특별자치도

for row in data :
    if name in row[0] :
        for i in row[3:104] : 
            i = i.replace(',', '')
            male.append(-int(i))        # 남자는 왼쪽화면에 출력
            m = m + int(i)
            
        for i in row[106:] :
            i = i.replace(',', '')
            female.append(int(i))       # 여자는 오른쪽 화면에 출력
            f = f + int(i)
            
        break                          # 두번째 데이터부터는 가져오지 않는다. 제주특별자치도로 시작하는 행이 많음

print('남자합계: ', m)
print('여자합계: ', f)
            
# ------------------------------------------------------------------------

print('남자 : ', len(male))
print('여자 : ', len(female))

plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi = 300)        # 그래프 크기 선택(inch)

plt.rc('font', family='Malgun Gothic')        # 맥 : Apple gothic  : 한글깨짐 방지
plt.rcParams['axes.unicode_minus'] = False   # - 기호 깨짐 방지

plt.title(name + '지역 남여 성별 비율')

plt.barh(range(101), male, label='남자')
plt.barh(range(101), female, label='여자')

plt.legend()                                  # 범례 출력 (label 선언이 선행되어야 함)
plt.show()
```

### 멧플롯립에서 사용할 수 있는 다양한 색상 출력

```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols

fig, ax = plt.subplots(figsize=(9, 8), dpi = 300)

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    row = i % nrows
    col = i // nrows
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(10),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```
