2. ndarray.size : 배열의 요소수를 확인한다.
3. ndarray.shape : 배열의 요소수를 정수형 튜플로 확인한다.

#### 슬라이싱의 규칙

1. 범위는 콜론기호를 사용한다.
2. 시작값 : 마지막값 형식으로 사용한다.
3. 시작값을 지정하면 지정된 위치부터 시작한다.
4. 시작값을 생략하면 0부터 시작한다.
5. 마지막값을 지정하면 <지정값 -1> 을 하여 진행한다.
6. 마지막값을 생략하면 마지막까지 진행한다.

### reshape(-1, 정수) : 행의 값에 -1을 사용하는 경우

```python
a = np.arange(12); a  # 1 * 12
```

```python
a.reshape(-1, 1)  # 12 * 1
```

```python
a.reshape(-1, 2)  # 6 * 2
```

```python
a.reshape(-1, 3)  # 4 * 3
```

```python
a.reshape(-1, 4)  # 3 * 4
```

```python
a.reshape(-1, 5)  # 5 * ? 요소 개수가 맞지 않아서 에러 발생
```

```python
a.reshape(-1, 6)  # 2 * 6
```
