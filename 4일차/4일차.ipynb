# 엑셀 다루기 : Xlsxwriter 모듈 사용하기

### 작업 순서 

1. import xlsxwriter 모듈 호출


2. 생성할 엑셀 워크북 객체 생성
    - workbook = xlsxwriter.Workbook(엑셀파일명)


3. 워크북 내에 사용할 워크시트 객체 생성
    - worksheet = workbook.add_worksheet([worksheet이름])


4. 워크 시트내에 작성할 셀 지정

    1) worksheet.write(row, col, cell_data)     # 셀의 행과 열의 위치를 지정
    
    2) worksheet.write(cell_address, cell_data) # 셀 주소로 지정(A4)


5. 워크북 객체를 닫고 엑셀 파일 생성
    - workbook.close()
    

# 실습1 : test1.xlsx

# 실습 : 다나와 30페이지 검색

```python
# 검색페이지를 반환하는 함수

def get_search_page_url(keyword, page):
    return 'http://search.danawa.com/dsearch.php?query={}&volumeType=allvs&page={}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&recommendedSort=Y&defaultUICategoryCode=112757&defaultPhysicsCategoryCode=860%7C13735%7C14883%7C58972&defaultVmTab=27307&defaultVaTab=1267793&tab=goods'.format(keyword, page)
```

```python
# 상품정보를 수집하는 함수

def get_prod_items(prod_items):
    prod_data = []

    for prod_item in prod_items:
        # 상품명 가져오기
        try:
            title = prod_item.select('p.prod_name > a')[0].text.strip()
        except:
            title = ''

        # 스펙정보 가져오기
        try:
            spec_list = prod_item.select('div.spec_list')[0].text.strip()
        except:
            spec_list = ''

        # 가격정보 가져오기
        try:
            price = prod_item.select('p.price_sect > a > strong')[0].text.strip().replace(',', '')
        except:
            price = 0

        prod_data.append([title, spec_list, price])
    
    return prod_data
```

```python
# 다나와 크롤링

from selenium import webdriver
from bs4 import BeautifulSoup
import time
from tqdm.notebook import tqdm


driver = webdriver.Chrome('chromedriver.exe')

keyword = '모니터'
total_page = 50

prod_data_total = []
for page in tqdm(range(1, total_page + 1)):
    # 검색페이지 이동
    url = get_search_page_url(keyword, page)
    driver.get(url)
    time.sleep(5)

    # 홈페이지 정보 분석
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')   # lxml 분석기(pip로 설치)
    
    # 상품정보 수집
    prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
    prod_item_list = get_prod_items(prod_items)
    
    # 수집된 정보 저장
    prod_data_total = prod_data_total + prod_item_list
    
    
driver.close()

print(len(prod_data_total))
print('-----------------------------------------------------------------------------------------------')
print(prod_data_total)
```

### 크롤링 된 정보를 엑셀로 저장하기

```python
import pandas as pd

data = pd.DataFrame(prod_data_total)
data.columns = ['상품명', '상품정보', '상품가격']
data.to_excel('./data/danawa_{}_crawling_result.xlsx'.format(keyword), index = False)
```

- 흡입력 단위를 표준화 한다.

```python
def get_suction(value):
    try:
        value = value.upper()
        if "AW" in value or "W" in value:
            result = value.replace("A", "").replace("W","")
            result = int(result.replace(",",""))
        elif "PA" in value:
            result = value.replace("PA","")
            result = int(result.replace(",",""))/100
        else:
            result = None
        return result
    except:
        return None
```

```python
new_suction_list = []

for power in suction_list:
    value = get_suction(power)
    new_suction_list.append(value)
```

- 사용시간을 분단위로 표준화하기

```python
def convert_time_minute(time):
    try:
        if '시간' in time:
            hour = time.split('시간')[0]
            if '분' in time:
                minute = time.split('시간')[-1].split('분')[0]
            else:
                minute = 0
        else: 
            hour = 0
            minute = time.split('분')[0]
        return int(hour)*60 + int(minute)
    except:
        return None
```

```python
times = ['40분', '4분', '1시간', '3시간 30분', '4시간']

for time in times:
    time_value = convert_time_minute(time)
    print(time, '=', time_value)
```

```python
# 제품별 사용시간을 분 단위로 표준화

new_use_time_list = []

for time in use_time_list:
    value = convert_time_minute(time)
    new_use_time_list.append(value)
```
