# glob 모듈

#### glob 모듈을 사용하여 파일이 위치한 폴더의 경로를 지정한다.

```python
#%%writefile d:/python_data/excel/excel_sys.py

import pandas as pd
from glob import glob
import sys

# sys 모듈을 사용하여 사용자에게 값을 입력받는다.
# input() 함수와 같다.
# sys.argv[0] 은 현재 작성중인 스크립트 파일을 의미한다.
# 사용법
# cmd 창에서 실핸한다.
# cmd > python 스크립트파일명 사용자입력값

input_dirs = sys.argv[1]

excel_data_files = glob(input_dirs)

total_data2 = pd.DataFrame()

for i in excel_data_files:
    df = pd.read_excel(i)
    total_data2 = total_data2.append(df, ignore_index = True)
    
print(total_data2)

total_data2.to_excel('데이터_통합.xlsx', index=False)
print()
print("파일이 저장되었습니다.")
```

# 담당자 변경

1. A -> BTS
2. B -> 뉴이스트
3. C -> IU

### 주의

파이썬은 원본파일을 직접 수정하지 않고 복사본을 사용하여 다른이름으로 저장하는 방식을
사용하고 있다.

```python
import pandas as pd
import re
from glob import glob

# 만약 glob를 import만 사용할 경우에는 glob.glob 해주어야 한다.

excel_data_files3 = glob('./excel/담당자*.xlsx')
#excel_data_files3

for i in excel_data_files3:
    df = pd.read_excel(i)
    
    if df.loc[1, '담당자'] == 'A':
        df['담당자'] = 'BTS'
        
    elif df.loc[1, '담당자'] == 'B':
        df['담당자'] = '뉴이스트'
        
    elif df.loc[1, '담당자'] == 'C':
        df['담당자'] = 'IU'
        
    i_new = re.sub('.xlsx', '2.xlsx', i)
    
    new_excel_file = pd.ExcelWriter(i_new, engine='xlsxwriter')
    df.to_excel(new_excel_file, index=False)
    new_excel_file.save()
    
# 작업완료된 파일들을 확인한다.
glob('./excel/담당자*.xlsx')
```

# 엑셀에 차트 삽입

```python
import pandas as pd
import re               # 정규표현식
import matplotlib       # 시각화
from glob import glob
import sys

# 통합 저장할 엑셀파일명 지정
excel_file_name = '담당자별_판매량_통합.xlsx'

# 판다스를 이용하여 통합 저장할 객체 생성
excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine = 'xlsxwriter')

# glob 모듈을 이용하여 파일이 있는 디렉토리를 스캔해준다.
excel_data_files = glob('담당자별_판매량_*사원.xlsx')

# 판다스의 데이터프레임 객체 생성
total_data = pd.DataFrame()

# 반복문을 이용하여 엑셀파일들을 저장한다.
for i in excel_data_files:
    df = pd.read_excel(i)
    total_data = total_data.append(df, ignore_index = True)
    
# 엑셀파일에 데이터 작성
total_data.to_excel(excel_total_file_writer, index = False, sheet_name = '통합문서')

# 차트 생성을 위한 워크북 객체 생성(엑셀파일 선택)
workbook = excel_total_file_writer.book

# 워크북내의 워크 시트 지정 -> 엑셀파일안에 새로운 시트이름을 지정
worksheet = workbook.add_worksheet('차트')

# 차트 생성 -> 선형타입으로 지정
chart1 = workbook.add_chart({'type': 'line'})

# 분기별 시계 판매량 / 담당자별로 집계
# andy
chart1.add_series({'values': '=통합문서!$D$2:$G$2',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$2'              # 사원지정
                 })

# becky
chart1.add_series({'values': '=통합문서!$D$5:$G$5',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$5'              # 사원지정
                 })

# chris
chart1.add_series({'values': '=통합문서!$D$8:$G$8',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$8'              # 사원지정
                 })


# 차트제목, x 축 이름, y 축 이름
chart1.set_title({'name' : '분기별 시계 판매량'})
chart1.set_x_axis({'name' : '분기'})
chart1.set_y_axis({'name' : '판매량'})


chart2 = workbook.add_chart({'type': 'line'})

# 분기별 구두 판매량 / 담당자별로 집계
chart2.add_series({'values': '=통합문서!$D$3:$G$3',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$2'              # 사원지정
                 })

# becky
chart2.add_series({'values': '=통합문서!$D$6:$G$6',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$5'              # 사원지정
                 })

# chris
chart2.add_series({'values': '=통합문서!$D$9:$G$9',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$8'              # 사원지정
                 })

# 차트제목, x 축 이름, y 축 이름
chart2.set_title({'name' : '분기별 구두 판매량'})
chart2.set_x_axis({'name' : '분기'})
chart2.set_y_axis({'name' : '판매량'})


chart3 = workbook.add_chart({'type': 'line'})

# 분기별 핸드백 판매량 / 담당자별로 집계
chart3.add_series({'values': '=통합문서!$D$4:$G$4',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$2'              # 사원지정
                 })

# becky
chart3.add_series({'values': '=통합문서!$D$7:$G$7',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$5'              # 사원지정
                 })

# chris
chart3.add_series({'values': '=통합문서!$D$10:$G$10',     # 실제 데이터
                  'categories': '=통합문서!$D$1:$G$1', #  헤더를 지정
                 'name': '=통합문서!$B$8'              # 사원지정
                 })

# 차트제목, x 축 이름, y 축 이름
chart3.set_title({'name' : '분기별 핸드백 판매량'})
chart3.set_x_axis({'name' : '분기'})
chart3.set_y_axis({'name' : '판매량'})


# 차트 삽입 위치를 지정한다.
worksheet.insert_chart('A2', chart1)
worksheet.insert_chart('I2', chart2)
worksheet.insert_chart('Q2', chart3)

# 데이터 저장 후 종료
excel_total_file_writer.save()
```
